Importing libraries

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

Loading Dataset

```python
data = pd.read_csv("/content/creditcard.csv")
data.head(25838)
```

Data Analysis

```python
data.info()
```

```python
data.describe()
```

Checking null/missing values

```python
data.isnull().sum()
```

```python
data.isnull()
```

Standardizing Time & Amount features 

```python
Time_mean=data['Time'].mean(axis=0)
Time_std=data['Time'].std(axis=0)
data['Time']-=Time_mean
data['Time']/=Time_std

Amount_mean = data['Amount'].mean(axis=0)
Amount_std=data['Amount'].std(axis=0)
data['Amount']-=Amount_mean
data['Amount']/=Amount_std
```

Separating labels from features

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, :-1].values
```

Splitting data 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state= 0)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size =0.3, random_state=0)
```

Modeling

```python
from keras import models
from keras import layers
from tensorflow.keras.regularizers import l2

model = models.Sequential()
model.add(layers.Dense(16, kernel_regularizer=l2(0.01), activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(16, kernel_regularizer=l2(0.01), activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

Compilation

```python
model.compile(optimizer='rmsprop',
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=4, batch_size=20, validation_data=(x_val, y_val))
```

Evaluation

```python
results = model.evaluate(x_test,y_test)
results
```
